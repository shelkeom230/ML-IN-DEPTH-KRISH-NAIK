# Missing Values

```python
# % of nan values in each feature
features_with_na=[feature for feature in df.columns if df[feature].isnull().sum()>1]

# print the feature name and % of missing values
for feature in features_with_na:
    print(feature,np.round(df[feature].isna().mean(),4),' % of missing values')
```

# As the dataset has many missing values, we need to check whether those missing values affect the target variable Sales Price
- Let's do some visualizations for this relationship

```python
for feature in features_with_na:
    data=df.copy()
    # replace the observations with null values by 1 else 0 

    data[feature]=np.where(data[feature].isnull(),1,0)

    # show the relationship of each feature having null values with SalePrice 
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

# Numerical Features

```python
numerical_features=[feature for feature in df.columns if df[feature].dtypes!='O']

print(f'Total numerical features: {len(numerical_features)}')

df[numerical_features].head()
```

# Temporal Variables (e.g Datetime variables)
we have some features with datetime values and we can use them to get some insights. Like , we can extract date from year and find out the difference between the year when house was built and it was sold

```python
# find out features with date time values 
year_features=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_features
```

```python
for feature in year_features:
    print(feature,df[feature].unique())
```

```python
for feature in year_features:
    print(feature,df[feature].unique().shape)
```

```python
# Let's analyse the temporal date time features 
# we will figure it out that , if there is a diff the year house is sold and the SalePrice

df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('Year Sold vs Median House Price')
```

```python
# Here, we will compare the difference between all year feature and SalePrice 
for feature in year_features:
    if feature!='YrSold':
        data=df.copy()

        # find out the diff between year variable and year the house was sold for 
        data[feature]=data['YrSold']-data[feature]

        plt.scatter(data[feature],data['YrSold'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
        
```
