```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
df=fetch_california_housing()
```

```python
df
```

```python
# create a dataframe
dataset=pd.DataFrame(df.data,columns=df.feature_names)
dataset.head()
```

```python
# add target
dataset['Price']=df.target
dataset.head()
```

```python
# divide the dataset into features and target variables
X=dataset.drop('Price',axis=1)
y=dataset['Price']
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
# cross validation
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
mean_mse
```

```python
lin_reg.fit(X_train,y_train)
```

```python
lin_reg.predict(X_test)
```

```python
# ridge regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()

params={'alpha':[1e-15,1e-13,1e-10,1e-8,1e-5,1e-3,1e-2,0,2,3,5,10,20,30,35,45,50,60,70,80,90,95,100]}

ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
(ridge_regressor.best_score_)
```

```python
# lasso regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

ridge=Lasso()

params={'alpha':[1e-15,1e-13,1e-10,1e-8,1e-5,1e-3,1e-2,0,2,3,5,10,20]}

lasso_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X_train,y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
# get r2_score
from sklearn.metrics import r2_score
y_pred=lasso_regressor.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# get r2_score for ridge
y_pred=ridge_regressor.predict(X_test)
r2_score(y_test,y_pred)
```

# Logistic Regression on breast cancer dataset

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
```

```python
dfnew=pd.DataFrame(df.data,columns=df.feature_names)
dfnew.head()
```

```python
dfnew['Cancer']=df.target
```

```python
dfnew.head()
```

```python
dfnew.Cancer.value_counts()
```

```python
# target varaible is balanced , perform train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(dfnew.drop('Cancer',axis=1),dfnew['Cancer'],test_size=0.33,random_state=42)
```

```python
log_reg=LogisticRegression(C=200,max_iter=1000)
```

```python
params=[{'C':[1,10,20,40,50]},{'max_iter':[500,1000,1100,1200,1250]}]
```

```python
model=GridSearchCV(log_reg,params,scoring='f1',cv=5)
model.fit(X_train,y_train)
```

```python
# accuracy score
from sklearn.metrics import  accuracy_score,
accuracy_score(y_test,log_reg.predict(X_test))
```

```python
print(model.best_params_)
print(model.best_score_)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
confusion_matrix(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

```python
accuracy_score(y_pred,y_test)
```
