```python
plt.plot(x,y,'r--',linestyle='dashed',linewidth=2,marker='+',markersize=12)
```

```python
x=np.arange(1,11)
y=3*x+5 
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("straight line equation")
```

```python
# plot a sine wave 
x=np.arange(0,5*m.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
y=np.cos(x)
plt.plot(x,y)
```

```python
x = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 500)
y=np.tan(x)
plt.plot(x,y,'g')
```

```python
x=np.arange(0,5*m.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

# create a subplot with 2 rows and 1 col and set it as active 
plt.subplot(2,1,1)
plt.title('sin graph')
plt.plot(x,y_sin,'r-')

# make the 2nd plot active 
plt.subplot(2,1,2)
plt.title('cos graph')
plt.plot(x,y_cos,'g-')
```

# Bar plot

```python
x=[10,11,13]
y=[20,30,40]

x2=[15,17,19]
y2=[33,35,36]

plt.bar(x,y)
plt.bar(x2,y2)
```

# Histplot

```python
arr=np.random.normal(loc=0,scale=1,size=1000)
plt.hist(arr,bins=20,density=True,edgecolor='black')
plt.xlabel('value')
plt.ylabel('freq')
plt.title('histogram freq distribution')
plt.grid(True)
```

# Box plot

```python
arr=[np.random.normal(0,std,100) for std in range(4)]
plt.boxplot(arr,vert=False,patch_artist=True) #patch_artist is for color
```

# Pie Chart 

```python
roles= ['SDE',' Data Scientist',' Data Analyst',' Business Analyst']
percentage_count=[40,20,30,10]
colors =['blue','red','green','purple']
explode=[0,0,0.1,0]
plt.pie(percentage_count,labels=roles,colors=colors,explode=explode,autopct='%1.1f%%')
```
