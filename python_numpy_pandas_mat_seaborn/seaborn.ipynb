# Seaborn Tutorial

## Distribution plots

- distplot
- pairplot
- joinplot

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
df.dtypes
```

# Co-relation for feature selection

```python
df.corr(numeric_only=True)
```

# HeatMap for feature selection
it shows the feature relevance with respect to target in the form of color shades and helps to choose features 

```python
sns.heatmap(df.corr(numeric_only=True))
```

# Join Plot 
perform univariate analysis 

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='scatter')
```

# PairPlot
it considers more than 1 feature

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Dist Plot 
helps to get the distribution of feature columns

```python
sns.displot(x='tip',data=df,kind='hist',kde=True)
```

```python
sns.displot(x='tip',data=df,kind='kde') # kde is kernel density estimator 
```
