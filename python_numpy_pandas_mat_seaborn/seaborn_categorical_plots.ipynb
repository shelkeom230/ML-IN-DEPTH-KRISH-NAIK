# Seaborn Categorical Varibles plot
- Count Plot
- Voilen Plot
- Box Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head()
```

# Count Plot
- shows frequency of each category for a feature

```python
sns.countplot(x='sex',data=df,color='green') #default label is taken as x axis 
```

```python
sns.countplot(x='smoker',data=df,color='red')
```

```python
sns.countplot(x='day',data=df,color='purple')
```

```python
sns.countplot(x='time',data=df,color='orange')
```

# Bar Plot 
we need to specify both x and y axis columns

```python
sns.barplot(x='sex',y='total_bill',data=df,color='indigo')
```

```python
sns.barplot(x='smoker',y='total_bill',data=df,color='yellow')
```

```python
sns.barplot(x='day',y='total_bill',data=df,color='pink')
```

```python
sns.barplot(x='sex',y='tip',data=df,color='orange')
```

```python
sns.barplot(x='smoker',y='tip',data=df,color='cyan')
```

# Box Plot
- shows 5 number summary

```python
sns.boxplot(x='sex',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(x='smoker',y='tip',data=df,palette='rainbow')
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
# consider entire data, also, boxplot works for both numeric and categorical columns
sns.boxplot(data=df,orient='h')
```

```python
# categorise data further based on some other features 
sns.boxplot(x='total_bill',y='day',data=df,hue='smoker',palette='dark:cyan')
```

# Violin Plot
it combines both box plot and KDE function, it shows both data distribution using 
KDE and box plot inside that 

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='dark:orange')
```
